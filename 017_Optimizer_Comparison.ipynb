```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

def rosenbrock(x, y):
    a = 1
    b = 100
    return (a - x) ** 2 + b * (y - x ** 2) ** 2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-2, 2, 100)
y = np.linspace(-1, 3, 100)
X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)

ax.plot_surface(X, Y, Z)

plt.show()
```

```python
# Create variables to optimize
x = tf.Variable(-1.0, trainable=True)
y = tf.Variable(-1.0, trainable=True)
x_value = []
y_value = []

SGD = tf.keras.optimizers.SGD(learning_rate=0.01)

for _ in range(4):
    with tf.GradientTape() as tape:
        # Compute the loss (which is your bivariate function)
        loss = rosenbrock(x, y)

    # Compute gradients
    gradients = tape.gradient(loss, [x, y])

    # Apply gradients to update variables
    SGD.apply_gradients(zip(gradients, [x, y]))
    x_value.append(x.numpy())
    y_value.append(y.numpy())

# Final values of x and y
print("Final x:", x.numpy())
print("Final y:", y.numpy())
```
